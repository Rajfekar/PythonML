```python
import gensim
import os
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

import nltk
nltk.download('punkt')

story = []
for filename in os.listdir('/content/ramayan'):
    # Check if it's a regular file and not a directory
    if os.path.isfile(os.path.join('/content/ramayan', filename)):
        try:
            # Try reading with 'utf-8' first, as it's more widely compatible.
            with open(os.path.join('/content/ramayan', filename), encoding='utf-8') as f:
                corpus = f.read()
        except UnicodeDecodeError:
            # If 'utf-8' fails, fallback to 'cp1252' or another appropriate encoding.
            # Here I use 'latin-1' as another option which also can handles special characters.
            # You might want to test 'iso-8859-1' if 'latin-1' does not work..
            # Consider 'utf-16' or other relevant encoding
            with open(os.path.join('/content/ramayan', filename), encoding='latin-1') as f:
                corpus = f.read()

        raw_sent = sent_tokenize(corpus)
        for sent in raw_sent:
            story.append(simple_preprocess(sent))
```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=10)
```
